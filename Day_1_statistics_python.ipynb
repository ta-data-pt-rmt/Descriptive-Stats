```
import pandas as pd
```

```
superstore = pd.read_excel("/content/Sample - EU Superstore.xls")
superstore.head()
```

```
# descriptive statistics
# take all the numerical columns and performs the basic statistics indicators on them
superstore.describe()
```

```
# you can also call aggregating functions on the whole dataframe
# returns a dataseries with the average value of each numerical column
superstore.mean()
```

```
# understand the relationship between sales and profit
superstore.plot.scatter(x = "Sales", y = 'Profit')
```

```
# seeing relationships through a scatter plot, is not sustainable, its vague
# how can we check the correlations between variables in a dataframe
superstore.corr()

# that correlation is a number that goes from -1 to 1
# correlation = 1 -> very strong positive correlation -> if one variable increases, so does the other
# correlation = 0 -> no correlation
# correlation = -1 -> very strong negative corrlation -> if one variable increases, the other decreases
```

```
# box plots -> great way to perform one dimensional analysis of a metric
superstore.plot.box()
```
